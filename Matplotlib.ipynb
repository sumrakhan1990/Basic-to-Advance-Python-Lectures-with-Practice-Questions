### Outline Matplotlib: From Basics to Advanced

#### Part 1: Introduction to Matplotlib
1. **Overview of Matplotlib**
   - What is Matplotlib?
   - Installation and setup
   - Importing Matplotlib

2. **Basic Plotting**
   - Creating simple plots (line plots, scatter plots, bar plots)
   - Customizing plots (titles, labels, legends)
   - Saving plots


```python
# matplotlib_introduction.py

import matplotlib.pyplot as plt
import numpy as np

# 1. Overview of Matplotlib
# Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.
# It is particularly useful for creating plots and graphs, ranging from simple line plots to complex 3D plots.

# Installation and setup
# To install Matplotlib, you can use pip:
# !pip install matplotlib

# Importing Matplotlib
# The most common way to import Matplotlib is to import the pyplot module as plt
# import matplotlib.pyplot as plt

# 2. Basic Plotting

# Creating simple plots

# Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='sin(x)')
plt.title('Line Plot of sin(x)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.show()

# Scatter Plot
x = np.random.rand(100)
y = np.random.rand(100)

plt.figure(figsize=(10, 5))
plt.scatter(x, y, c='red', marker='o')
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()

# Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 2]

plt.figure(figsize=(10, 5))
plt.bar(categories, values, color='blue')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

# Customizing plots

# Adding Titles, Labels, and Legends
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('Line Plot of sin(x) and cos(x)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.show()

# Saving plots

# Saving a plot to a file
plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='sin(x)')
plt.title('Line Plot of sin(x)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.savefig('line_plot.png')
plt.show()

# Example of combining multiple plots into a single figure using subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

# First subplot: Line plot
axes[0].plot(x, y1, label='sin(x)')
axes[0].set_title('Line Plot of sin(x)')
axes[0].set_xlabel('x-axis')
axes[0].set_ylabel('y-axis')
axes[0].legend()
axes[0].grid(True)

# Second subplot: Scatter plot
axes[1].scatter(x, y2, c='red', marker='o')
axes[1].set_title('Scatter Plot of cos(x)')
axes[1].set_xlabel('x-axis')
axes[1].set_ylabel('y-axis')
axes[1].grid(True)

plt.show()
```


1. **Overview of Matplotlib**:
   - Matplotlib is a powerful plotting library for Python.
   - It can be installed using `pip install matplotlib`.
   - The `pyplot` module from Matplotlib is commonly imported as `plt`.

2. **Basic Plotting**:
   - **Line Plot**: Created using `plt.plot()`, with customization options for title, labels, legend, and grid.
   - **Scatter Plot**: Created using `plt.scatter()`, with customization options for color and markers.
   - **Bar Plot**: Created using `plt.bar()`, useful for categorical data representation.

3. **Customizing Plots**:
   - Titles, labels, and legends are added using `plt.title()`, `plt.xlabel()`, `plt.ylabel()`, and `plt.legend()`.
   - Plots can be saved to files using `plt.savefig()`.

4. **Subplots**:
   - Multiple plots can be combined into a single figure using `plt.subplots()`.
   - Each subplot can be individually customized.

4. **Multiple Line Plots in a Single Figure**:

   - Create a simple plot of the function \( y = \sin(x) \) for \( x \) values ranging from 0 to \( 2\pi \).

```python
### Solutions for Practice Problems

import matplotlib.pyplot as plt
import numpy as np

# 1. Simple Line Plot
x = np.linspace(0, 10, 100)
y = 2 * x + 3

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='y = 2x + 3')
plt.title('Line Plot of y = 2x + 3')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.legend()
plt.show()

# 2. Customizing a Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(10, 5))
plt.scatter(x, y, c='blue', marker='o', label='Random Points')
plt.title('Scatter Plot of Random Points')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()

# 3. Bar Plot with Custom Colors
categories = ['E', 'F', 'G', 'H']
values = [8, 3, 6, 9]

plt.figure(figsize=(10, 5))
plt.bar(categories, values, color='green')
plt.title('Bar Plot of Categories')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

# 4. Multiple Line Plots in a Single Figure
x = np.linspace(-10, 10, 200)
y1 = x ** 2
y2 = x ** 3

plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='y = x^2', color='red', linestyle='--')
plt.plot(x, y2, label='y = x^3', color='blue', linestyle='-')
plt.title('Line Plots of y = x^2 and y = x^3')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.show()

# 5. Saving a Plot to a File
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='sin(x)')
plt.title('Sine Wave')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.savefig('sine_wave.png')
plt.show()

# 6. Subplots
x1 = np.linspace(0, 2 * np.pi, 100)
y1 = np.cos(x1)
x2 = np.linspace(-np.pi / 4, np.pi / 4, 100)
y2 = np.tan(x2)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

# First subplot: Cosine plot
axes[0].plot(x1, y1, label='cos(x)')
axes[0].set_title('Cosine Plot')
axes[0].set_xlabel('x-axis')
axes[0].set_ylabel('y-axis')
axes[0].legend()
axes[0].grid(True)

# Second subplot: Tangent plot
axes[1].plot(x2, y2, label='tan(x)', color='orange')
axes[1].set_title('Tangent Plot')
axes[1].set_xlabel('x-axis')
axes[1].set_ylabel('y-axis')
axes[1].legend()
axes[1].grid(True)

plt.show()
```

```python
# matplotlib_basic_plotting.py

import matplotlib.pyplot as plt
import numpy as np

# 1. Line Plots
# Creating a simple line plot

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='sin(x)')
plt.title('Line Plot of sin(x)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.show()

# Customizing line plots with different styles and colors
plt.figure(figsize=(10, 5))
plt.plot(x, y, 'r--', label='sin(x)')  # Red dashed line
plt.title('Customized Line Plot of sin(x)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.show()

# 2. Scatter Plots
# Creating a simple scatter plot

x = np.random.rand(100)
y = np.random.rand(100)

plt.figure(figsize=(10, 5))
plt.scatter(x, y, c='blue', marker='o', label='Data points')
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.show()

# Customizing scatter plots with different colors and markers
plt.figure(figsize=(10, 5))
plt.scatter(x, y, c='green', marker='^', label='Data points')  # Green triangles
plt.title('Customized Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.show()

# 3. Bar Plots
# Creating a simple bar plot

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 4]

plt.figure(figsize=(10, 5))
plt.bar(categories, values, color='blue')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

# Customizing bar plots with different colors and edge colors
plt.figure(figsize=(10, 5))
plt.bar(categories, values, color='orange', edgecolor='black')
plt.title('Customized Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()

# 4. Histograms
# Creating a simple histogram

data = np.random.randn(1000)

plt.figure(figsize=(10, 5))
plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Customizing histograms with different bin sizes and colors
plt.figure(figsize=(10, 5))
plt.hist(data, bins=50, color='green', alpha=0.7, edgecolor='black')
plt.title('Customized Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# 5. Subplots
# Creating multiple subplots in a single figure

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

# First subplot: Line plot of sin(x)
axes[0].plot(x, y1, label='sin(x)', color='blue')
axes[0].set_title('Line Plot of sin(x)')
axes[0].set_xlabel('x-axis')
axes[0].set_ylabel('y-axis')
axes[0].legend()
axes[0].grid(True)

# Second subplot: Line plot of cos(x)
axes[1].plot(x, y2, label='cos(x)', color='red')
axes[1].set_title('Line Plot of cos(x)')
axes[1].set_xlabel('x-axis')
axes[1].set_ylabel('y-axis')
axes[1].legend()
axes[1].grid(True)

plt.show()

# Customizing subplots with different layouts and spacing
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

# Top-left subplot: Line plot of sin(x)
axes[0, 0].plot(x, y1, label='sin(x)', color='blue')
axes[0, 0].set_title('Line Plot of sin(x)')
axes[0, 0].set_xlabel('x-axis')
axes[0, 0].set_ylabel('y-axis')
axes[0, 0].legend()
axes[0, 0].grid(True)

# Top-right subplot: Line plot of cos(x)
axes[0, 1].plot(x, y2, label='cos(x)', color='red')
axes[0, 1].set_title('Line Plot of cos(x)')
axes[0, 1].set_xlabel('x-axis')
axes[0, 1].set_ylabel('y-axis')
axes[0, 1].legend()
axes[0, 1].grid(True)

# Bottom-left subplot: Scatter plot
axes[1, 0].scatter(x, y1, label='sin(x) points', color='green')
axes[1, 0].set_title('Scatter Plot of sin(x)')
axes[1, 0].set_xlabel('x-axis')
axes[1, 0].set_ylabel('y-axis')
axes[1, 0].legend()
axes[1, 0].grid(True)

# Bottom-right subplot: Histogram
axes[1, 1].hist(data, bins=30, color='purple', alpha=0.7, edgecolor='black')
axes[1, 1].set_title('Histogram')
axes[1, 1].set_xlabel('Value')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()
```

1. **Line Plots**:

2. **Scatter Plots**:
   - A scatter plot is created using `plt.scatter()`.

```python
import matplotlib.pyplot as plt
import numpy as np

# Task 1: Line Plot Customization
x = np.linspace(-10, 10, 100)
y = x**2

plt.figure(figsize=(10, 5))
plt.plot(x, y, 'g-', label='x^2')
plt.title('Line Plot of x^2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.show()

# Task 2: Scatter Plot Customization
x = np.random.rand(100)
y = np.random.rand(100)

plt.figure(figsize=(10, 5))
plt.scatter(x, y, c='blue', marker='s', label='Data points')
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.grid(True)
plt.show()

# Task 3: Bar Plot Customization
categories = ['Apple', 'Banana', 'Cherry', 'Date']
values = [10, 15, 7, 12]

plt.figure(figsize=(10, 5))
plt.bar(categories, values, color=['red', 'yellow', 'pink', 'brown'])
plt.title('Bar Plot of Fruits')
plt.xlabel('Fruit')
plt.ylabel('Quantity')
plt.show()

# Task 4: Histogram Customization
data = np.random.randn(1000)

plt.figure(figsize=(10, 5))
plt.hist(data, bins=20, color='purple', alpha=0.7, edgecolor='black')
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Task 5: Creating Subplots
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
data_random = np.random.rand(100)
data_uniform = np.random.uniform(-1, 1, 1000)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

# Top-left subplot: Line plot of sin(x)
axes[0, 0].plot(x, y_sin, label='sin(x)', color='blue')
axes[0, 0].set_title('Line Plot of sin(x)')
axes[0, 0].set_xlabel('x-axis')
axes[0, 0].set_ylabel('y-axis')
axes[0, 0].legend()
axes[0, 0].grid(True)

# Top-right subplot: Line plot of cos(x)
axes[0, 1].plot(x, y_cos, label='cos(x)', color='red')
axes[0, 1].set_title('Line Plot of cos(x)')
axes[0, 1].set_xlabel('x-axis')
axes[0, 1].set_ylabel('y-axis')
axes[0, 1].legend()
axes[0, 1].grid(True)

# Bottom-left subplot: Scatter plot
axes[1, 0].scatter(data_random, data_random, label='Random points', color='green')
axes[1, 0].set_title('Scatter Plot of Random Points')
axes[1, 0].set_xlabel('x-axis')
axes[1, 0].set_ylabel('y-axis')
axes[1, 0].legend()
axes[1, 0].grid(True)

# Bottom-right subplot: Histogram
axes[1, 1].hist(data_uniform, bins=30, color='purple', alpha=0.7, edgecolor='black')
axes[1, 1].set_title('Histogram of Uniformly Distributed Data')
axes[1, 1].set_xlabel('Value')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
import matplotlib.gridspec as gridspec
from matplotlib.patches import FancyArrowPatch
import matplotlib.patches as patches
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
from matplotlib.backends.backend_agg import FigureManager

# Part 5: Advanced Techniques and Optimization

# 1. Subplots and Complex Layouts

# Creating Subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
x = np.linspace(0, 10, 100)
y = np.sin(x)

axs[0, 0].plot(x, y, 'r')
axs[0, 0].set_title('Subplot 1')
axs[0, 1].plot(x, np.cos(x), 'g')
axs[0, 1].set_title('Subplot 2')
axs[1, 0].scatter(x, y, c='b')
axs[1, 0].set_title('Subplot 3')
axs[1, 1].bar(x[:10], np.random.randint(1, 10, size=10))
axs[1, 1].set_title('Subplot 4')
plt.tight_layout()
plt.show()

# Customizing Subplot Layouts
fig = plt.figure(figsize=(10, 7))
gs = gridspec.GridSpec(3, 3, width_ratios=[1, 2, 1], height_ratios=[2, 1, 1])

ax0 = plt.subplot(gs[0, :])
ax1 = plt.subplot(gs[1, :-1])
ax2 = plt.subplot(gs[1:, -1])
ax3 = plt.subplot(gs[2, :-1])

ax0.plot(x, y)
ax1.scatter(x, y)
ax2.bar(x[:10], np.random.randint(1, 10, size=10))
ax3.plot(x, np.cos(x), 'g')

ax0.set_title('Top plot')
ax1.set_title('Left bottom plot')
ax2.set_title('Right bottom plot')
ax3.set_title('Bottom plot')
plt.show()

# 2. Animation

# Creating Basic Animations
fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'r-', animated=True)

def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(np.sin(frame))
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128),
                    init_func=init, blit=True)
plt.show()

# Customizing Animations
fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'b-', animated=True)

def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(np.sin(frame))
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128),
                    init_func=init, blit=True, interval=50,
                    repeat=True, repeat_delay=1000)
plt.show()

# Saving Animations as GIFs or Videos
fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'r-', animated=True)

def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(np.sin(frame))
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 128),
                    init_func=init, blit=True)
ani.save('animation.mp4', writer='ffmpeg')  # Save as MP4 video
# ani.save('animation.gif', writer='imagemagick')  # Uncomment to save as GIF

# 3. Performance Optimization

# Handling Large Datasets
n = 1000000
x = np.linspace(0, 100, n)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y, lw=0.5)
ax.set_title('Handling Large Datasets')
plt.show()

# Optimizing Rendering Performance
fig, ax = plt.subplots()
x = np.linspace(0, 10, 1000)
y = np.sin(x)

# Use blitting for faster updates
line, = ax.plot(x, y, animated=True)

def update(frame):
    line.set_ydata(np.sin(x + frame / 10.0))
    return line,

ani = FuncAnimation(fig, update, frames=100, blit=True, interval=50)
plt.show()
```
